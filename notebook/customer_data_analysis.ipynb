## 2- Clean Data

- customers_01 :

```python
customers_01 = customers.drop(columns= ["Unnamed","  Gender  "])
```

```python
customers_01.sample(5)
```

- customers_02 :

```python
customers_02 = customers_01.drop_duplicates()
```

```python
customers_02.duplicated().sum()
```

```python
customers_02["Gender"].unique()
```

```python
customers_02.sample(5)
```

- customers_03 :

```python
customers_03 = customers_02.copy()
customers_03["Gender"] = customers_03["Gender"].str.upper()
```

```python
customers_03["Gender"].unique()
```

```python
customers_03["Gender"] = customers_03["Gender"].replace({"FEMALE": "Female", "MALE": "Male","F": "Female", "M": "Male"})
```

```python
customers_03["Gender"].unique()
```

```python
customers_03.head(25)
```

```python
customers_03["Gender"].value_counts(dropna=False)
```

```python
customers_03.head(25)
```

- customers_04 :

```python
customers_04 = customers_03.copy()
customers_04["Gender"] = customers_04["Gender"].fillna("Male")
```

```python
customers_04["Gender"].unique()
```

```python
customers_04.head(25)
```

- customers_05 :

```python
customers_05 = customers_04.copy()
customers_05["Rating"] = customers_04["Rating"].apply(lambda x: np.nan if x < 1 else x)
customers_05["Rating"] = customers_04["Rating"].apply(lambda x: customers_05["Rating"].mean() if x > 5 else x)
```

```python
customers_05["Age"] = customers_05["Age"].apply(lambda x: x if x >= 7 and x < 100 else customers_05["Age"].median())
```

- customers_06 :

```python
customers_06 = customers_05.copy()
customers_06["ProductCategory"] = customers_05["ProductCategory"].fillna("No Category Selected")
```

```python
customers_06.info()
```

```python
customers_06["ProductCategory"].unique()
```

```python
customers_06.head(25)
```

```python
customers_05["Age"].describe()
```

```python
customers_06["PurchaseAmount"].describe()
```

- customers_07 :

```python
customers_07 = customers_06.copy()
customers_07["PurchaseAmount"] = customers_06["PurchaseAmount"].fillna(customers_06["PurchaseAmount"].mean())
```

```python
customers_07.isna().sum()
```

```python
customers_07.info()
```

```python
customers_07["PurchaseDate"] = pd.to_datetime(customers_07["PurchaseDate"], errors="coerce")
```

```python
customers_07["PurchaseDate"].head(25)
```

```python
customers_07.nunique()
```

```python
customers_07["Phone"].value_counts(dropna=False)
```

```python
customers_07["Phone"] = customers_07["Phone"].apply(lambda x: "Available" if pd.notnull(x) else "Missing")
```

```python
customers_07.info()
```

- final customers :

```python
final_customers = customers_07.copy()
```

## 3- Data Analysis

```python
final_customers.sort_values(by="PurchaseAmount", ascending=False).head(5)
```

```python
final_customers.sort_values(by=["PurchaseAmount","Rating"], ascending=True).head(5)
```

```python
final_customers["Rating"].value_counts().sort_index(ascending=True)
```

```python
Gender_counts = final_customers["Gender"].value_counts()
```

```python

plt.pie(Gender_counts,labels=["Males","Females"],autopct="%1.1f%%", startangle=90,colors=("#72dcff", "#fb7ae1"), explode=(0.005,0.005),textprops={"fontsize": 20, "color": "black"},)
plt.title("Distribution of Customers by Gender", fontsize=25, color="black")
plt.show()
```

```python
sns.histplot(final_customers["Age"], bins=10,kde=True, color="#A644EC")
plt.title("Distribution of Customers by Age", fontsize=20, color="Black")
plt.xlabel("Age", fontsize=10)
plt.ylabel("Frequency", fontsize=10)
plt.show()
```

```python
sns.boxplot(x="Rating", y="ProductCategory", data=final_customers,color="#c2d50f")
plt.title("Distribution of Ratings by Product Category", fontsize=20, color="black")
plt.show()
```

```python
Correlation = final_customers[["Age","Rating","PurchaseAmount"]].corr().round(4)
```

```python
sns.heatmap(Correlation, annot=True, cmap="coolwarm", fmt=".4f",linewidths=0.5, cbar_kws={"shrink": 0.8}, square=True, annot_kws={"size": 12})
plt.title("Correlation Matrix", fontsize=20, color="black")
plt.show()
```

```python
sns.scatterplot(x="Age", y="PurchaseAmount", data=final_customers, hue="Gender", palette={"Male": "#72dcff", "Female": "#fb7ae1"},)
plt.title("Scatter Plot of Age vs Purchase Amount by Gender", fontsize=20, color="black")
plt.show()
```

```python
sns.barplot(x="Gender", y="PurchaseAmount", data=final_customers, hue="Gender", palette={"Male": "#72dcff", "Female": "#fb7ae1"},estimator=np.mean)
plt.title("Bar Plot of Purchase Amount by Gender", fontsize=20, color="black")  
plt.show()  
```

```python
final_customers["Months"] = final_customers["PurchaseDate"].dt.month_name()
month_order = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]
monthly_purchases = final_customers.groupby("Months")["PurchaseAmount"].sum().reindex(month_order)
```

```python
monthly_purchases.plot(kind="line", color="#00E43D", linewidth=2, figsize=(15, 4),marker="o", markersize=6, markerfacecolor="#8AFF41", markeredgewidth=1, markeredgecolor="black")
plt.title("Monthly Purchases", fontsize=20, color="black")
plt.xlabel("Months", fontsize=14)
plt.ylabel("Total Purchase Amount", fontsize=14)
plt.xticks(rotation=45)
plt.show()
```
